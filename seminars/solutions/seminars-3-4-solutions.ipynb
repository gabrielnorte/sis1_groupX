```python
# exercise 6

import matplotlib.pyplot as plt
import numpy as np

fs = 10000
T0 = .1
t = np.arange (-T0*2, T0*2, 1/fs)
tp = np.arange (0, T0, 1/fs)
x = np.zeros (int(T0*fs), dtype=float)
x[0:int(.25*T0*fs)] = 1.0
x[int(.75*T0*fs):int(T0*fs)] = 1.0
x = np.append(x,x)
x = np.append(x,x)
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(t,x, linewidth=2)
plt.axis([-.2,.2,-.1,1.1])
plt.xlabel('time (sec)')
plt.grid()
plt.ylabel('amplitude')
```

```python
# exercise 7

import matplotlib.pyplot as plt
import numpy as np

fs = 10000
T0 = 8
t = np.arange (-T0, T0, 1/fs)
tp = np.arange (0, T0, 1/fs)
x = np.zeros (int(2*T0*fs))
x[int(.5*T0*fs):int(T0*fs)] = 4+np.arange (-4, 0, 1/fs)
x[int(1.5*T0*fs):int(2*T0*fs)] = 4+np.arange (-4, 0, 1/fs)
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(t,x, linewidth=2)
plt.axis([-8,8,-.1,4.1])
plt.xlabel('time (sec)')
plt.grid()
plt.ylabel('amplitude')
```
