```python
# exercise 1.a

import matplotlib.pyplot as plt
import numpy as np

plt.figure(1, figsize=(9.5, 3.5))
plt.vlines([1,4], [0,0], [1,-2], linewidth=2)
plt.axis([0,5,-2.5,1.5])
plt.xlabel('time (n)')
plt.grid()
plt.ylabel('amplitude')
```
