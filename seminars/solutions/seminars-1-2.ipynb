```python
# exercise 1

import matplotlib.pyplot as plt
import numpy as np

omega = np.pi/5
fs = 10000
t = np.arange (-10, 20, 1/fs)
x = 3 * np.cos(omega*t - (np.pi/3))
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(t, x, linewidth=2)
plt.axis([-10,20,-3.5,3.5])
plt.xlabel('time (sec)')
plt.grid()
plt.ylabel('amplitude')
```

```python
# exercise 2.a

import matplotlib.pyplot as plt
import numpy as np

fs = 10000
phase = np.arange (0, 6*np.pi, 1/fs)
x = np.cos(phase)
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(phase, x, linewidth=2)
last=6*np.pi
plt.axis([0, last, -1.1, 1.1])
plt.xlabel('phase')
plt.grid()
plt.ylabel('amplitude')
```

```python
# exercise 3

import matplotlib.pyplot as plt
import numpy as np

A = 9
omega = 2*np.pi * (20/9)
phi = -2*np.pi/9
fs = 10000
t = np.arange (-.5, .5, 1/fs)
x = A*np.cos(omega*t+phi)
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(t, x, linewidth=2)
plt.axis([-.5, .5, -10, 10])
plt.xlabel('time (sec)')
plt.grid()
plt.ylabel('amplitude')
```

```python
# exercise 11

import matplotlib.pyplot as plt
import numpy as np

A = np.sqrt(65)
omega = 10*np.pi
phi = np.arctan(-4/7)
fs = 10000
t = np.arange (-1, 1, 1/fs)
x = A*np.cos(omega*t+phi)
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(t, x, linewidth=2)
plt.axis([-1, 1, -A-1, A+1])
plt.xlabel('time (sec)')
plt.grid()
plt.ylabel('amplitude')
```

```python
import matplotlib.pyplot as plt
import numpy as np
X=2*np.exp(1j*np.pi/4)
z=0.9*np.exp(1j*np.pi/8)
n = np.arange(100)
plt.figure(1, figsize=(9.5, 3.5))
plt.plot(np.real(X*np.power(z,n)), label='Real part')
plt.axis([0, 50, -1.8, 1.8])
plt.plot(np.imag(X*np.power(z,n)), label='Imaginary part')
plt.axis([0, 50, -1.8, 1.8])
plt.xlabel ('n')
plt.legend()
```
